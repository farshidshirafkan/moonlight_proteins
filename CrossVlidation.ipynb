```python
import  pandas  as  pd
import  numpy   as  np
import  seaborn as  sb
import  matplotlib.pyplot as plt
from sklearn.preprocessing      import    StandardScaler  
from  sklearn.naive_bayes       import    GaussianNB
from  sklearn.neighbors         import    KNeighborsClassifier
from  sklearn.svm               import    SVC
from  sklearn.neural_network    import    MLPClassifier
from  sklearn.tree              import    DecisionTreeClassifier
from  sklearn.ensemble          import    RandomForestClassifier
from  sklearn.ensemble          import    AdaBoostClassifier
from  sklearn.linear_model      import    LogisticRegression
from  sklearn.model_selection   import    cross_val_score
from sklearn.metrics            import    accuracy_score
from sklearn.metrics            import    precision_score
from sklearn.metrics            import    recall_score
from sklearn.metrics            import    f1_score
from sklearn.metrics            import    roc_auc_score
###
class  Moonlihght_CV():                                     
    def FiletoDF(self, path):
        f = pd.read_csv(path)
        df= pd.DataFrame(f) 
        y = df['target'] 
        t = df.drop(['uniprot'], axis=1)
        X = t.drop(['target'], axis=1)
        return  X ,y  
    
    def CreateModel(self):
        NB  = GaussianNB()        
        KNN = KNeighborsClassifier()  
        SVM = SVC()   
        MLP = MLPClassifier()       
        DT  = DecisionTreeClassifier()
        RF  = RandomForestClassifier() 
        ADA = AdaBoostClassifier()
        LR  = LogisticRegression()
        return [NB, KNN, SVM, MLP, DT, RF, ADA, LR]    
       
    def CrossValidation(self, X, y, score, k):
        lst = self.CreateModel()
        NBS  = round(np.mean(cross_val_score(lst[0], X, y, scoring=score, cv=k)), 2)    
        KNNS = round(np.mean(cross_val_score(lst[1], X, y, scoring=score, cv=k)), 2)           
        SVMS = round(np.mean(cross_val_score(lst[2], X, y, scoring=score, cv=k)), 2)    
        MLPS = round(np.mean(cross_val_score(lst[3], X, y, scoring=score, cv=k)), 2)
        DTS  = round(np.mean(cross_val_score(lst[4], X, y, scoring=score, cv=k)), 2)    
        RFS  = round(np.mean(cross_val_score(lst[5], X, y, scoring=score, cv=k)), 2)    
        ADAS = round(np.mean(cross_val_score(lst[6], X, y, scoring=score, cv=k)), 2)    
        LRS  = round(np.mean(cross_val_score(lst[7], X, y, scoring=score, cv=k)), 2)
        return NBS, KNNS, SVMS, MLPS, DTS, RFS, ADAS, LRS    
      
    def cvd(self, X, y, k, mes):       
        cv = self.CrossValidation(X, y, mes, k)
        model = ['NB', 'KNN', 'SVM', 'MLP', 'DT', 'RF', 'ADA', 'LR']
        df = pd.DataFrame(cv, index=model) #, columns=lst
        return df
    
    def main(self, X, y, k):
        dfacc = self.cvd(X, y, k, 'accuracy' )
        dfpre = self.cvd(X, y, k, 'precision')
        dfrec = self.cvd(X, y, k, 'recall'   )
        dff1  = self.cvd(X, y, k, 'f1'       )
        dfauc = self.cvd(X, y, k, 'roc_auc'  )
        dfacc = dfacc.rename(columns={0: 'Accuracy'})
        dfpre = dfpre.rename(columns={0: 'Precision'})
        dfrec = dfrec.rename(columns={0: 'Recall'})
        dff1  = dff1.rename(columns= {0: 'F1'})
        dfauc = dfauc.rename(columns={0: 'AUC'})
        result = pd.concat([dfacc.T, dfpre.T, dfrec.T, dff1.T, dfauc.T])
        return result

###

obj = Moonlihght_CV()
X, y = obj.FiletoDF('D:/GitHub-IBB/DATA/QSorder.csv')
k = 10
result = obj.main(X, y, k)
result
```
