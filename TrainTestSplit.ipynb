```python
import  pandas            as    pd
import  numpy             as    np
import  seaborn           as    sb
import  matplotlib.pyplot as    plt
from  sklearn.model_selection   import  train_test_split
from  sklearn.preprocessing     import  StandardScaler  
from  sklearn.naive_bayes       import  GaussianNB
from  sklearn.neighbors         import  KNeighborsClassifier
from  sklearn.svm               import  SVC
from  sklearn.neural_network    import  MLPClassifier
from  sklearn.tree              import  DecisionTreeClassifier
from  sklearn.ensemble          import  RandomForestClassifier
from  sklearn.ensemble          import  AdaBoostClassifier
from  sklearn.linear_model      import  LogisticRegression
from  sklearn.metrics           import  accuracy_score
from  sklearn.metrics           import  precision_score
from  sklearn.metrics           import  recall_score
from  sklearn.metrics           import  f1_score
from  sklearn.metrics           import  roc_auc_score
from  sklearn.metrics           import  matthews_corrcoef
###

class  TrainTestSplitClass(): 
    def FiletoDF(self, path):
        f = pd.read_csv(path)
        df= pd.DataFrame(f) 
        t = df.drop(['uniprot'], axis=1)
        X = t.drop(['target'], axis=1)
        y = df['target']
        return X ,y   
        
    def DoFiletoDF(self, path):
        X=[0]
        y=[0]
        X[0], y[0] = self.FiletoDF(path)       
        return X, y  
    
    def CreateModel(self):
        NB  = GaussianNB()        
        KNN = KNeighborsClassifier()  
        SVM = SVC()   
        MLP = MLPClassifier()       
        DT  = DecisionTreeClassifier()
        RF  = RandomForestClassifier() 
        ADA = AdaBoostClassifier()
        LR  = LogisticRegression()
        return [NB, KNN, SVM, MLP, DT, RF, ADA, LR] 
    
    def TrainTestSplit(self, X, y):
        Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)
        return Xtrain, Xtest, ytrain, ytest

    def Standardization(self, Xtrain, Xtest):   
        scaler = StandardScaler()
        scaler.fit(Xtrain) 
        Xtrain = scaler.transform(Xtrain)  
        Xtest  = scaler.transform(Xtest)  
        return Xtrain, Xtest

    def FitPredict(self, Xtrain, ytrain, Xtest):
        Model = self.CreateModel()       
        Model[0].fit(Xtrain, ytrain)
        Model[1].fit(Xtrain, ytrain)
        Model[2].fit(Xtrain, ytrain)
        Model[3].fit(Xtrain, ytrain) 
        Model[4].fit(Xtrain, ytrain)
        Model[5].fit(Xtrain, ytrain)
        Model[6].fit(Xtrain, ytrain)  
        Model[7].fit(Xtrain, ytrain)  
        ypNB  = Model[0].predict(Xtest)
        ypKNN = Model[1].predict(Xtest)
        ypSVM = Model[2].predict(Xtest)  
        ypMLP = Model[3].predict(Xtest)
        ypDT  = Model[4].predict(Xtest)
        ypRF  = Model[5].predict(Xtest)
        ypADA = Model[6].predict(Xtest)
        ypLR  = Model[7].predict(Xtest)
        yPred = [ypNB, ypKNN, ypSVM, ypMLP, ypDT, ypRF, ypADA, ypLR]
        return yPred 
       
    def aprfrm(self, ytest, yPredlst, funscore):
        NB  = round(funscore(ytest, yPredlst[0]), 2)    
        KNN = round(funscore(ytest, yPredlst[1]), 2)             
        SVM = round(funscore(ytest, yPredlst[2]), 2)    
        MLP = round(funscore(ytest, yPredlst[3]), 2)  
        DT  = round(funscore(ytest, yPredlst[4]), 2)    
        RF  = round(funscore(ytest, yPredlst[5]), 2)   
        ADA = round(funscore(ytest, yPredlst[6]), 2)     
        LR  = round(funscore(ytest, yPredlst[7]), 2)     
        my_Score = [ NB, KNN, SVM, MLP, DT, RF, ADA, LR ]
        return my_Score  
  
    def ComputeScore(self, pXtrain, pXtest, pytrain, pytest):
        sXtrain, sXtest = self.Standardization(pXtrain, pXtest)
        yPred_model = self.FitPredict(sXtrain, pytrain, sXtest)       
        acc_all = self.aprfrm(pytest, yPred_model, accuracy_score)
        pre_all = self.aprfrm(pytest, yPred_model, precision_score)
        rec_all = self.aprfrm(pytest, yPred_model, recall_score)
        f1_all  = self.aprfrm(pytest, yPred_model, f1_score)
        auc_all = self.aprfrm(pytest, yPred_model, roc_auc_score)
        mcc_all = self.aprfrm(pytest, yPred_model, matthews_corrcoef)
        return  acc_all, pre_all, rec_all, f1_all, auc_all, mcc_all
    
    def DoComputeScore(self, pXtrain, pXtest, pytrain, pytest):
        acc =[0]; pre =[0]; rec =[0]; f1m =[0] ; auc =[0] ;  mcc =[0]
        acc[0], pre[0], rec[0], f1m[0], auc[0], mcc[0] = self.ComputeScore(pXtrain[0], pXtest[0], pytrain[0], pytest[0])
        return  acc, pre, rec, f1m, auc, mcc         

    def  TTS(self, X, y):
        Xtrain=[0]; Xtest=[0]; ytrain=[0] ; ytest=[0]
        Xtrain[0], Xtest[0], ytrain[0], ytest[0] = self.TrainTestSplit(X[0], y[0])       
        return  Xtrain, ytrain, Xtest, ytest

    def MD(self, acc, pre, rec, f1, auc, mcc):
        Models = ['NB', 'KNN', 'SVM', 'MLP', 'DT', 'RF', 'ADA', 'LR']
        a = pd.DataFrame(acc, index=['Accuracy'] , columns=Models)
        b = pd.DataFrame(pre, index=['Precision'], columns=Models)
        c = pd.DataFrame(rec, index=['Recall']   , columns=Models)
        d = pd.DataFrame(f1 , index=['F1']       , columns=Models)
        e = pd.DataFrame(auc, index=['AUC']      , columns=Models)
        f = pd.DataFrame(mcc, index=['MCC']      , columns=Models)  
        return a, b, c, d, e, f
    
    
    def  Main(self, X, y):
        Xtrain, ytrain, Xtest, ytest = self.TTS(X, y)
        acc, pre, rec, f1, auc, mcc = self.DoComputeScore(Xtrain, Xtest, ytrain, ytest)
        dfacc, dfpre, dfrec, dff1, dfauc, dfmcc = self.MD(acc, pre, rec, f1, auc, mcc)
        result = pd.concat([dfacc, dfpre, dfrec, dff1, dfauc, dfmcc]).T
        return result
###

obj = TrainTestSplitClass()   
X, y = obj.DoFiletoDF('D:/GitHub-IBB/DATA/QSorder.csv')
result = obj.Main(X, y)
result
```
